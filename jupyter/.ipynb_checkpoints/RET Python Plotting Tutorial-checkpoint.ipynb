## Plotting in Python

### Basics

The most commonly used plotting library is the pyplot module in the matplotlib library, we'll only be using this library for the basic plottings.

First thing you want to do is to import it, the "as" keywords basically allows you to give it a nick name so you don't have to type the whole thing again and again. It's the same as importing numpy.

```python
import matplotlib.pyplot as plt
# this way anytime you use plt, it'll be the same as calling matplotlib.pyplot
# of course you can name it whatever you want, it's a variable name, that represents the "matplotlib.pyplot" thingy.
```

The basic functionsfor plotting that you'll be using, is "plot", "scatter" and maybe "errorbar".

plt.plot( ) for connected plots (the points will be connected by line segments):

```python
x_coord = [0,1,2,3,4]
y_coord = [0,1,2,3,4]

plt.plot(x_coord,y_coord)
plt.show()
```

plt.scatter( ) for scatter point plots:

```python
plt.scatter(x_coord,y_coord)
plt.show()
```

plt.errorbar( ), allows you to add a list for error bar values to the plot in the third input:

```python
# The multiplication here works the same way as string x integer
error = [0.2]*len(x_coord)

plt.errorbar(x_coord,y_coord,error)
plt.show()
```

### Input options and keywords

The 3 plotting functions above also have other input options, so you can change the line color, line width, marker shape etc. You can find these options in their documentations here:

plot( ): https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.plot.html

scatter( ): https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.scatter.html

errorbar( ): https://matplotlib.org/api/_as_gen/matplotlib.pyplot.errorbar.html

Take plot( ) for example, these are some common options' keywords:
1. color, or c. Input as string. See plot() documentation above for available choices.
1. linestyle, or ls. Input as string. See plot() documentation above for available choices.
1. linewidth or lw. Input as float or int number.
1. marker or m. Input as string. See plot() documentation above for available choices.
1. markersize or ms. Input as float or int number.
1. label, a piece of text that'll display in the legends, if you opt to plot legends. We'll come back to this later.

There's also a 'fmt' option keyword, which can be a combination of color and linestyle for simplicity.

```python
# The below 3 ways to enter inputs are equivalent and will produce the same plot
# The last one is the most abbreviated format, which is most commonly used.

plt.plot( x_coord, y_coord, color = 'r',linestyle = '--', linewidth = 1.0, marker = 'D', markersize = 5.0 )
plt.show()
plt.plot( x_coord, y_coord, c = 'r', ls = '--', lw = 1.0, marker = 'D', ms = 5.0 )
plt.show()
plt.plot( x_coord, y_coord,'r--', lw = 1.0, marker = 'D', ms = 5.0 )
plt.show()
```

You can search for other options in the plot( ) documentations. Note that some options are distinct to plot(), the keywords can be different in scatter( ) and errorbar( ), you'll need to check the documentations for sure.

There's also other common plotting functions such as hist( ) for histogram, you can find others here:
https://matplotlib.org/3.1.0/api/pyplot_summary.html


### Utility functions
There are other functions that can help you modify your plot. The above 3 plotting functions only creates the lines and dots for the plot, that's it, to make your plot more presentable there are other functions you can use.

We already used one in the above examples, the show( ) function. Which will display the plot. Although in jupyter notebook, even if you don't use show( ), it'll output it for you. This is not true when you're doing this on a terminal.

```python
plt.plot(x_coord,y_coord)  # On jupyter notebook you don't need to explicitly use show() on plt.
plt.show()
```

Commonly used utility functions:
1. xlabel( ), ylabel( ), title( ), legend( ), text( )
2. xticks( ), yticks( )
3. xlim( ), ylim( )
4. show( ), savefig( ), clf( ), figsize( )

If you want to use these, please Google their documentations, there's too many different options for me to cover here, they aren't vital to your plot, as the actual lines and dots are the essense, they just help make the plot professional and organized, here's an example though:

```python
import numpy as np
mass = [i for i in range(40,101)]
radius = [1.2*A**(1.0/3) for A in mass]
# Set figure size, you need to do this before using the plotting function, i.e. the line of plt.scatter( ) 
plt.figure(figsize=(10,5))

# Create the lines and/or dots with plotting functions
plt.scatter(mass,radius,color = 'k',s = 1, label = "radius")

# Create x-axis notation
plt.xlabel("Mass Number A", fontsize = 20)

# Create y-axis notation
plt.ylabel("Radius (fm)", fontsize = 20)

# Create title
plt.title("Radius vs Mass Number", fontsize = 20)

# Create legend, useful when you have many sets of data
plt.legend(loc = 'upper left',fontsize = 20)

# Put a text message somewhere on the graph, the first 2 inputs are the x,y coordinate of the start of the message
plt.text(80,4.4,"pokemon",fontsize = 15)

# Define limits for x and y axis, sometimes you might want to extend a little to include legends or text etc.
# If you don't use these functions, Python will adjust automatically.
plt.xlim(30,120)
plt.ylim(1,6)

# Define the tick marks you want to plot, you can even change it to text if you want
# First argument is the list for locations to put the ticks, second argument is optional, 
# it should be a list of text that has the same dimension as the location list.
# If you don't use these functions, Python will adjust automatically.
plt.xticks([40,50,60],['monkey','lion','elephant'], fontsize=15)
plt.yticks(np.arange(1,6,0.5))

# save figure, I commented this out so it doesn't create a saved file everytiand cause a mess.
plt.savefig("python_plotting_tutorial.pdf",format = 'pdf')

# Display the plot or figure, this should be placed after savefig(), otherwise savefig() will save a blank figure.
plt.show()

# Clear the plot or figure, useful when you have a loop and you are making lots of plots, if you don't 
# clear it, it'll be drawn on top of each other instead of individual plots.
plt.clf()


```

Other things to keep in mind:

1. The x and y coordinate input lists need to have the same length, same goes for error bar (the 3rd input in errorbar( ) ).
 
2. The data points in your plot are defined by index, meaning x[0] and y[0] together defines a point, so before plotting, you need to create these two lists carefully so that for index i, x[i] and y[i] come from the same data point.

3. The ordering of points in scatter( ) doesn't matter because you are only plotting points (you still need to have x[i] and y[i] coming from the same data point). But in plot( ) and errorbar( ) , the ordering decides how the line segments are connected, thus the ordering is important. See demonstration below:

```python
# Plot quadratic function
x = [1,2,3,4,5]
y = [i**2 for i in x]
plt.plot(x,y,'r-',marker = 'o',ms=5)

# f"" is format string type, you can put a pair of curly brakets {} inside a string and the inside of it will be
# executed as codes. You don't need to learn this right now, just showing the neat tricks.
print (f"Plotting points {[(x[i],y[i]) for i in range(len(x))]}")  
plt.show()

# Rotate the ordering of x, the index with colon here is called "slicing" in Python, 
# you don't have to learn how to use this right now.
x_rotate = x[3:] + x[:3] 
y_rotate = y[3:] + y[:3]

print (f"Plotting points {[(x_rotate[i],y_rotate[i]) for i in range(len(x))]}")  
plt.plot(x_rotate,y_rotate,'r-',marker = 'o',ms=5)
plt.show()
```

See how (5,25) and (1,1) are connected because they're next to each other in the input you gave plot( )? And that (3,9) and (4,16) aren't connected because (3,9) is at the end of the list and it's connected to nothing. The way plot( ) determine which points to connect is determined by the ordering, it always connects the neighboring points only. 

Obviously scatter( ) doesn't care because there are no line segments.

### Plotting multiple sets of data

Plotting multiple sets of data is straight forward, you simply write down a series of plt.plot( ) before you do plt.show( ). 

Let's draw lines and some power functions on the same figure, together with some random scattered points.

```python
import random
# Just in case there's some leftover plots from earlier.
plt.clf()

x = list(range(5))
y = x
y2 = [i**2 for i in x]
y3 = [i**3 for i in x]
y_rand = [random.randint(0,125) for i in x]

plt.figure(figsize=(10,5))
plt.plot(x,y,'k-',label = 'linear', lw=1)
plt.plot(x,y2,'b--',label = 'quadratic')
plt.plot(x,y3,'r-.',label = 'cubic')
plt.scatter(x,y_rand,c='y', marker='^',label = 'random')

plt.legend(fontsize=12)
plt.show()

```
